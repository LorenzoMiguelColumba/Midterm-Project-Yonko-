### Objective 4

Finally, for the prediction model that will be created for this project, we will be utilizing the Markov chain algorithm. The Markov property states that the probability of transitioning to the next state depends only on the current state and is independent of the previous states.

$$(i)P(X_{n+1}=j \;|\; X_n= i)=P(i,j) $$

Equation i represents the statement of the markov property in which the probability of being in state j at the next time step (n+1) given that the current state is i (at time n) is equal to the transition probability P(i, j) [2]. Due to the demands of the Markov algorithm we need to sort the temperature data which is called as a State Space, this is defined as a set of all possible states that our temperatures can be sorted in, for our ranges we will sort the data as States: A ,B, and C for their ranges (-∞, 28), (29, 31), and (30, ∞+) respectively. After sorting the states the occurrences of every state change will be recorded as per equation i, the result will then be processed to probabilities and then will be distributed in a matrix.
